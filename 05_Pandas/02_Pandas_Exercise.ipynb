# Filtering DataFrame
* Filtering data means selecting only the rows that match certain conditions from a dataset.
* You keep what you need and remove what you dont.


```python
# find all the final winners
final = matches['MatchNumber'] == "Final"
new_df = matches[final]
new_df[['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
matches[matches['SuperOver'] == "Y"].shape[0]
```

```python
# how many matches has csk won in kolkata
kolkata_matches = matches[matches['City'] == 'Kolkata']
kolkata_matches[kolkata_matches['WinningTeam'] == 'Chennai Super Kings'].shape[0]

# 2nd method to write this code
matches[(matches['City'] == "Kolkata") & (matches['WinningTeam'] == "Chennai Super Kings")].shape[0]
```

```python
matches.head(2)
```

```python
# toss winner is match winner in percentage
total_matches = matches.shape[0]
print("Total number of matches: ",total_matches)

toss_winner_matches = matches[matches['TossWinner'] == matches['WinningTeam']]

toss_win_count = toss_winner_matches.shape[0]
print("Toss win count :", toss_win_count)

percentage = (toss_win_count/total_matches) * 100
print("Percentage: ",percentage)

# A short method to write this code
percentage = (matches[matches['TossWinner'] == matches['WinningTeam']].shape[0]/matches.shape[0]) * 100
print("Percentage by short code ",percentage)
```

```python
# movies with rating highier than 8 amdb and votes > 10000
required = movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
print(required)
```
